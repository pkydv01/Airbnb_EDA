````python
#lots of missing values found in last_review and reviews_per_month
#first going with last_review

df['last_review'].astype('datetime64[ns]')
df['last_review'].dtype
````

````python
# last review is like a categorical column and datatype as object, so replace with mode is better way to handle missing values
df['last_review'].mode()
````

````python
df['last_review']=df['last_review'].fillna('2019-06-23')     #filled last_review with mode
````

````python
#now to fill 'reviews_per_month' column let's check its mean,median,mode
print('mean value of reviews_per_month column: ',df['reviews_per_month'].mean())
print('median value of reviews_per_month column:',df['reviews_per_month'].median())
print('mode value of reviews_per_month column:', df['reviews_per_month'].mode())
````

````python
df['reviews_per_month'].fillna(df['reviews_per_month'].median(), inplace=True)    #used median imputation-->median is robust to outliers and works well for skewed data.
````

````python
# we have only 16 missing values in name and 21 missing values in host_name out of 48895 data-->very few, so impact is low.
#Drop rows (Best for modeling)-->Missing name likely makes the row unusable. So, I'm ready to drop these columns

df.dropna(subset=['name','host_name'], inplace=True)
````

````python
#checking if any more missing values are there-->
df.isnull().sum().sort_values(ascending=False).head()
````

````python
#Grouping and aggregating data is also a part of data manipulation. So, lets group some data to answer some questions

# What is the average price of listings in each neighbourhood group?
avg_price_by_neighbourhood_group = df.groupby('neighbourhood_group')['price'].mean().sort_values()
avg_price_by_neighbourhood_group
````

````python
# Which room type is the most common in each neighbourhood group?
common_room_type_by_neighbourhood_group = df.groupby('neighbourhood_group')['room_type'].value_counts()
common_room_type_by_neighbourhood_group
````

````python
# What is the average number of reviews per month for different room types?
avg_num_reviews_by_room_type = df.groupby('room_type')['reviews_per_month'].mean()
avg_num_reviews_by_room_type
````

````python
# How many listings have more than 100 reviews?

listings_with_more_than_100_reviews = df[df['number_of_reviews'] > 100]
listings_with_more_than_100_reviews.reset_index()
````

````python
# What is the correlation between price and minimum nights?
correlation_price_min_nights = df['price'].corr(df['minimum_nights'])
correlation_price_min_nights
````

````python
#Which neighbourhood has the highest average price?
avg_price_by_neighbourhood = df.groupby('neighbourhood')['price'].mean().idxmax()
avg_price_by_neighbourhood
````

````python
# What percentage of listings are available for more than 300 days in a year?
listings_available_more_than_300_days = df[df['availability_365'] > 300]
percentage_available_more_than_300_days = (len(listings_available_more_than_300_days) / len(df)) * 100
percentage_available_more_than_300_days
````

````python
# How many hosts have more than 5 listings?
hosts_with_more_than_5_listings = df.groupby('host_id').filter(lambda x: len(x) > 5)
num_hosts_with_more_than_5_listings = len(hosts_with_more_than_5_listings['host_id'].unique())
num_hosts_with_more_than_5_listings
````

````python
# What is the distribution of prices for entire home/apt vs private rooms?
price_distribution_room_type = df.groupby('room_type')['price'].describe()
price_distribution_room_type
````

````python
# What is the average availability of listings with more than 50 reviews?
avg_availability_with_more_than_50_reviews = df[df['number_of_reviews'] > 50]['availability_365'].mean()
avg_availability_with_more_than_50_reviews
````
